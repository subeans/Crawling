```python
from selenium import webdriver
from bs4 import BeautifulSoup
import urllib.request as ur
import random
import time
import json
```

```python
def getFeature(good_list):
    options=webdriver.ChromeOptions()
    driver=webdriver.Chrome('./chromedriver')
    driver.implicitly_wait(1)
    driver.get('https://www.amazon.com/s?i=fashion-mens-intl-ship&bbn=16225019011&rh=n%3A16225019011%2Cn%3A1040658%2Cn%3A1045558&dc&pf_rd_i=16225019011&pf_rd_m=ATVPDKIKX0DER&pf_rd_p=5cd8272b-5ce4-4c26-bfcb-d6dca0c1e427&pf_rd_p=5cd8272b-5ce4-4c26-bfcb-d6dca0c1e427&pf_rd_r=P4D7CAAS3T9F4CH9CJMT&pf_rd_r=P4D7CAAS3T9F4CH9CJMT&pf_rd_s=merchandised-search-left-2&pf_rd_t=101&qid=1563170684&rnid=1040658&ref=sr_nr_n_6')
    i=0
    id_list=[]
    while True:
        html=driver.page_source
        soup=BeautifulSoup(html,'html.parser')
        a_last=driver.find_elements_by_class_name('a-last')[0]
        a_last.click()
        ids=soup.findAll('div',attrs={'class':'sg-col-4-of-24 sg-col-4-of-12 sg-col-4-of-36 s-result-item sg-col-4-of-28 sg-col-4-of-16 sg-col sg-col-4-of-20 sg-col-4-of-32'})
        for j in range(48):
            id=ids[j].get('data-asin')
            print(id)
            id_list.append(id)
        driver.implicitly_wait(7)
        i+=1
        if (i==151):
            break;
    info={
        'id':id_list
    }
    
    good_list.append(info)
        
```

```python
good_list=[]
getFeature(good_list)
json.dump(good_list,open('./Pants_Id','w+'))
```
